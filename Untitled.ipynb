```python
import networkx as nx

G = nx.Graph()
G.add_nodes_from(['Mateo', 'Jose', 'Pablo', 'Carlos'], gender="M", node_color="tab:red")
G.add_nodes_from(['Daisy', 'Camila', 'Vanesa'], gender="F", color = 'blue')
G.add_nodes_from(['Cami'], gender="NB", color = 'green')


pos = nx.spring_layout(G, seed=3113794652)  # positions for all nodes

nx.draw_networkx_nodes(G,pos, nodelist=['Mateo', 'Cami', 'Pablo', 'Carlos','Daisy', 'Camila', 'Vanesa'], node_color="tab:grey")
nx.draw_networkx_nodes(G,pos, nodelist=['Jose'], node_color="tab:red")

nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[('Mateo', 'Cami'), ('Mateo', 'Pablo'), ('Pablo', 'Daisy'), ('Camila', 'Vanesa'), ('Vanesa', 'Carlos'),
             ('Pablo', 'Carlos'), ('Daisy', 'Cami')],
    width=2,
    alpha=0.5
)


nx.draw_networkx_edges(
    G,
    pos,
    edgelist=[('Mateo', 'Jose'), ('Jose', 'Pablo'), ('Jose', 'Daisy'), ('Jose', 'Vanesa'), ('Jose', 'Carlos'),
              ('Jose', 'Cami')],
    width=6,
    alpha=0.5,
    edge_color="tab:blue"
)
```
