```python
import networkx as nx

G = nx.Graph()
G.add_nodes_from(['Mateo', 'Jose', 'Pablo', 'Carlos'], gender="M")
G.add_nodes_from(['Daisy', 'Camila', 'Vanesa'], gender="F")
G.add_nodes_from(['Cami'], gender="NB")
G.add_edges_from([('Mateo', 'Jose'), ('Jose', 'Pablo'), ('Jose', 'Carlos')])
G.add_edges_from([('Daisy', 'Camila'), ('Camila', 'Vanesa'), ('Vanesa', 'Daisy')])
nx.draw(G, with_labels = True)
print(nx.attribute_assortativity_coefficient(G, "gender"))
```
